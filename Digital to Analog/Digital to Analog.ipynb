```python
#ASK
import numpy as np
import matplotlib.pyplot as plt

inp = input("Enter Digital Data:")

low_amplitude = 0
high_amplitude = 5
phase = 0
frequency = 1

signal = np.array([])
time = np.array([])
start = 0

for ch in inp:
    temporary_time = np.arange(start,start+2,0.001)
    if ch == '0':
        temporary_signal = low_amplitude*np.sin(2*np.pi*frequency*temporary_time+phase)
    else:
        temporary_signal = high_amplitude*np.sin(2*np.pi*frequency*temporary_time+phase)
    signal = np.append(signal,temporary_signal)
    time = np.append(time,temporary_time)
    start = start + 2

plt.grid()
plt.title("ASK")
plt.plot(time,signal,color='blue')
plt.show()
```

```python
#FSK
import numpy as np
import matplotlib.pyplot as plt

inp = input("Enter Digital Data:")

amplitude = 5
phase = 0
low_frequency = 1
high_frequency = 2

signal = np.array([])
time = np.array([])
start = 0

for ch in inp:
    temporary_time = np.arange(start,start+2,0.001)
    if ch == '0':
        temporary_signal = amplitude*np.sin(2*np.pi*low_frequency*temporary_time+phase)
    else:
        temporary_signal = amplitude*np.sin(2*np.pi*high_frequency*temporary_time+phase)
    signal = np.append(signal,temporary_signal)
    time = np.append(time,temporary_time)
    start = start + 2

plt.grid()
plt.title("FSK")
plt.plot(time,signal,color='cyan')
plt.show()
```

```python
#PSK
import numpy as np
import matplotlib.pyplot as plt

inp = input("Enter Digital Data:")

amplitude = 5
phase = 0
frequency = 1

signal = np.array([])
time = np.array([])
start = 0

for ch in inp:
    temporary_time = np.arange(start,start+2,0.001)
    if ch == '0':
        temporary_signal = amplitude*np.sin(2*np.pi*frequency*temporary_time+phase)
    else:
        temporary_signal = amplitude*np.sin(2*np.pi*frequency*temporary_time*(-1)+phase)
    signal = np.append(signal,temporary_signal)
    time = np.append(time,temporary_time)
    start = start + 2

plt.grid()
plt.title("PSK")
plt.plot(time,signal,color='violet')
plt.show()
```
