```python
#For Unipolar Representation
import numpy as np
import matplotlib.pyplot as plt

inp = input("Enter Digital Signal:")

signal = list()

for ch in inp:
    if ch == '0':
        signal.append(0)
    else:
        signal.append(1)

signal.append(0)
x = np.arange(0,len(signal))

plt.ylabel("Unipolar")
plt.grid()
plt.step(x,signal,where='post')

plt.show()
```

```python
#NRZ-L Representation
import numpy as np
import matplotlib.pyplot as plt

inp = input("Enter Digital Signal:")

signal = list()

for ch in inp:
    if ch == '0':
      signal.append(1)
    elif ch == '1':
      signal.append(-1)
    
signal.append(0)

x = np.arange(0,len(signal))

plt.ylabel("NRZ-L")
plt.grid()
plt.step(x,signal,where='post')
plt.show()
```

```python
#NRZ-I Representation

import numpy as np
import matplotlib.pyplot as plt

inp = input("Enter Digital Signal: ")

signal = list()

for ch in inp:
    if ch == '0':
        signal.append(0)
    else:
        signal.append(1)
        

for i in range(len(signal)):
    if signal[i] == 0 and i == 0:
        signal[i] = 1
    elif signal[i] == 1 and i>0:
        if signal[i-1] == 1:
            signal[i] = -1
        elif signal[i-1] == -1:
            signal[i] = 1
    elif signal[i] == 0 and i>0:
        signal[i] = signal[i-1]
        
signal.append(0)

x = np.arange(0,len(signal))
plt.ylabel("NRZ-I")
plt.grid()
plt.step(x,signal,where='post')
plt.show()
```

```python
#RZ Representation

import numpy as np
import matplotlib.pyplot as plt

def getSignal(sig):
    signal = list()
    
    for i in sig:
        if i == 0:
            signal.append(-1)
            signal.append(0)
        else:
            signal.append(1)
            signal.append(0)
            
    signal.append(0)
    return signal


inp = input("Enter Digital Signal: ")

sig = list()
signal = list()

for ch in inp:
    if ch == '0':
        sig.append(0)
    else:
        sig.append(1)

signal = getSignal(sig)

x = np.arange(0,len(signal))
plt.ylabel("RZ")
plt.grid()
plt.step(x,signal,color='blue',where='post')
plt.show()
```

```python
#Bi-Phase Manchester

import numpy as np
import matplotlib.pyplot

def getSignal(sig):
    signal = list()
    for i in sig:
        if i == 0:
            signal.append(1)
            signal.append(-1)
        else:
            signal.append(-1)
            signal.append(1)
    
    signal.append(0)
    return signal


inp = input('Enter Digital Signal: ')

sig = list()
signal = list()

for ch in inp:
    if ch =='0':
        sig.append(0)
    else:
        sig.append(1)

signal = getSignal(sig)

x = np.arange(0,len(signal))

plt.ylabel("Bi-Phase Manchester")
plt.grid()
plt.step(x,signal,where='post')
plt.show()
```

```python
#Bi-Phase Differential Manchester

import numpy as np
import matplotlib.pyplot as plt


def getSignal(sig):
    signal = list()

    if sig[0] == 0:
        signal.append(-1)
        signal.append(1)
    else:
        signal.append(1)
        signal.append(-1)
    
    for i in range(1,len(sig)):
        if sig[i] == 1:
            signal.append(signal[-1])
            signal.append(-1*signal[-2])
        else:
            signal.append(-1*signal[-1])
            signal.append(signal[-2])
            
    return signal
    
inp = input("Enter Digital Signal: ")

sig = list()
signal = list()

for ch in inp:
    if ch == '0':
        sig.append(0)
    else:
        sig.append(1)

signal = getSignal(sig)

x = np.arange(0,len(signal))

plt.ylabel("Bi-Phase Differential Manchester")
plt.grid()
plt.step(x,signal,color='red',where='post')
plt.show()
```
