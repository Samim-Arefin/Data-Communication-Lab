```python
#Unipolar Representation

import numpy as np
import matplotlib.pyplot as plt

inp = input("Enter Digital Data:")

signal = []

for ch in inp:
    if ch == '0':
        signal.append(0)
    else:
        signal.append(1)
        
signal.append(0)

time = np.arange(0,len(signal))

plt.grid()
plt.ylabel("Unipolar")
plt.step(time,signal,where='post')
plt.show()
```

```python
#NRZ-L Representation

import numpy as np
import matplotlib.pyplot as plt

inp = input("Enter Digital Data:")

signal = []

for ch in inp:
    if ch == '0':
        signal.append(1)
    else:
        signal.append(-1)
        
signal.append(0)
time = np.arange(0,len(signal))

plt.grid()
plt.ylabel("NRZ-L")
plt.step(time,signal,where='post')
plt.show()
```

```python
#NRZ-I Representation

import numpy as np
import matplotlib.pyplot as plt

def getSignal(sig):
    for i in range(1,len(sig)):
        if sig[i] == -1:
            if sig[i-1] == -1:
                sig[i] = 1
            elif sig[i-1] == 1:
                sig[i] = -1
        else:
            sig[i] = sig[i-1]
            
    return sig
     

inp = input("Enter Digital Data:")

sig = []

for ch in inp:
    if ch == '0':
        sig.append(1)
    else:
        sig.append(-1)
        
signal = getSignal(sig)
signal.append(0)
time = np.arange(0,len(signal))

plt.grid()
plt.ylabel("NRZ-I")
plt.step(time,signal,where='post')
plt.show()
```

```python
#RZ Representation

import numpy as np
import matplotlib.pyplot as plt

def getSignal(inp):
    signal = []
    for ch in inp:
        if ch =='0':
            signal.append(-1)
            signal.append(0)
        else:
            signal.append(1)
            signal.append(0)
      
    return signal

inp = input("Enter Digital Data:")

signal = getSignal(inp)
signal.append(0)

time = np.arange(0,len(signal))

plt.grid()
plt.ylabel("RZ")
plt.step(time,signal,where='post',color='violet')
plt.show()
```

```python
#Bi-Phase Manchester

import numpy as np
import matplotlib.pyplot as plt

def getSignal(inp):
    signal = []
    for ch in inp:
        if ch == '0':
            signal.append(1)
            signal.append(-1)
        else:
            signal.append(-1)
            signal.append(1)
      
    return signal

inp = input("Enter Digital Data:")

signal = getSignal(inp)
signal.append(0)

time = np.arange(0,len(signal))

plt.grid()
plt.ylabel("Bi-Phase Manchester")
plt.step(time,signal,where='post',color='purple')
plt.show()
```

```python
#Bi-Phase Differential Manchester
import numpy as np
import matplotlib.pyplot as plt

def getSignal(inp):
    signal = []
    if inp[0] =='0':
        signal.append(-1)
        signal.append(1)
    else:
        signal.append(1)
        signal.append(-1)
    
    for i in range(1,len(inp)):
        if inp[i] =='1':
            signal.append(signal[-1])
            signal.append(-1*signal[-2])
        else:
            signal.append(signal[-2])
            signal.append(-1*signal[-1])
            
    return signal

inp = input("Enter Digital Data:")

signal = getSignal(inp)
signal.append(0)
time = np.arange(0,len(signal))

plt.grid()
plt.ylabel("Bi-Phase Differential Manchester")
plt.step(time,signal,where='post',color='grey')
plt.show()
```
