```python
#For Unipolar Representation
import numpy as np
import matplotlib.pyplot as plt

inp = input("Enter Digital Signal:")

signal = list()

for ch in inp:
    if ch == '0':
        signal.append(0)
    elif ch == '1':
        signal.append(1)

signal.append(0)
x = np.arange(0,len(signal),1)

plt.ylabel("Unipolar")
plt.step(x,signal,where='post')

plt.show()
```

```python
#NRZ-I Representation
import numpy as np
import matplotlib.pyplot as plt

inp = input("Enter Digital Signal:")

signal = list()

for ch in inp:
    if ch == '0':
      signal.append(1)
    elif ch == '1':
      signal.append(-1)
    
signal.append(0)

x = np.arange(0,len(signal),1)

plt.ylabel("NRZ-I")
plt.step(x,signal,where='post')
plt.show()
```
