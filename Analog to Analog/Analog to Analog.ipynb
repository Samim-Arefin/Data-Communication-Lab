```python
#Amplitude Modulation
import numpy as np
import matplotlib.pyplot as plt

M_A = 3 #float(input("Enter Message Signal Amplitude"))
M_F = 5 #float(input("Enter Message Signal Frequency"))
C_A = 7 #float(input("Enter Carrier Signal Amplitude"))
C_F = 70 #float(input("Enter Carrier Signal Frequency"))
phase = 0
time = np.arange(0,1,0.001)

#Carrier Signal
Carrier_Signal = C_A*np.cos(2*np.pi*C_F*time+phase)

#Modulating Signal
Modulating_Signal = M_A*(np.sin(2*np.pi*M_F*time+phase)**3 - np.cos(2*np.pi*M_F*time+phase)**2)

#Modulated Signal
Modulated_Signal = (C_A+Modulating_Signal)*np.cos(2*np.pi*C_F*time+phase)

plt.subplot(3,1,1)
plt.title("Amplitude Modulation")
plt.grid()
plt.xlabel("Modulating Signal")
plt.ylabel("Amplitude")
plt.plot(time,Modulating_Signal,color='red')

plt.subplot(3,1,2)
plt.grid()
plt.xlabel("Carrier Signal")
plt.ylabel("Amplitude")
plt.plot(time,Carrier_Signal,color='blue')

plt.subplot(3,1,3)
plt.grid()
plt.xlabel("Modulated Signal")
plt.ylabel("Amplitude")
plt.plot(time,Modulated_Signal,color='black')

plt.subplots_adjust(hspace=1)
fig = plt.gcf()
fig.set_size_inches(8,9)
plt.show()
```

```python
#Frequency Modulation
import numpy as np
import matplotlib.pyplot as plt

M_A = 3 #float(input("Enter Message Signal Amplitude"))
M_F = 5 #float(input("Enter Message Signal Frequency"))
C_A = 7 #float(input("Enter Carrier Signal Amplitude"))
C_F = 10 #float(input("Enter Carrier Signal Frequency"))
phase = 0

time = np.arange(0,1,0.001)

#Carrier Signal
Carrier_Signal = C_A*np.sin(2*np.pi*C_F*time+phase)

#Modulating Signal
Modulating_Signal = M_A*np.sin(2*np.pi*M_F*time+phase)

#Modulated Signal
Modulated_Signal = C_A*np.sin(2*np.pi*(C_F*time+Modulating_Signal)+phase)

plt.subplot(3,1,1)
plt.title("Frequency Modulation")
plt.grid()
plt.xlabel("Modulating Signal")
plt.ylabel("Amplitude")
plt.plot(time,Modulating_Signal,color='red')

plt.subplot(3,1,2)
plt.grid()
plt.xlabel("Carrier Signal")
plt.ylabel("Amplitude")
plt.plot(time,Carrier_Signal,color='blue')

plt.subplot(3,1,3)
plt.grid()
plt.xlabel("Modulated Signal")
plt.ylabel("Amplitude")
plt.plot(time,Modulated_Signal,color='black')

plt.subplots_adjust(hspace=1)
fig = plt.gcf()
fig.set_size_inches(8,9)
plt.show()
```

```python
#Phase Modulation

import numpy as np
import matplotlib.pyplot as plt

M_A = 3 #float(input("Enter Message Signal Amplitude"))
M_F = 5 #float(input("Enter Message Signal Frequency"))
C_A = 7 #float(input("Enter Carrier Signal Amplitude"))
C_F = 30 #float(input("Enter Carrier Signal Frequency"))

time = np.arange(0,1,0.001)

#Modulating Signal
Modulating_Signal = M_A*np.sin(2*np.pi*M_F*time)
Converted_MSignal = Modulating_Signal.astype(int)
Carrier_Signal = C_A*np.sin(2*np.pi*C_F*time)

for value in Converted_MSignal:
    if(Converted_MSignal[value])>=0:
        phase = 0.5*Converted_MSignal
    else:
        phase = -0.5*(Converted_MSignal)
        
Modulated_Signal = C_A*np.sin(2*np.pi*C_F*time+phase)

plt.subplot(3,1,1)
plt.title("Phase Modulation")
plt.grid()
plt.xlabel("Modulating Signal")
plt.ylabel("Amplitude")
plt.plot(time,Modulating_Signal,color='red')

plt.subplot(3,1,2)
plt.grid()
plt.xlabel("Carrier Signal")
plt.ylabel("Amplitude")
plt.plot(time,Carrier_Signal,color='blue')

plt.subplot(3,1,3)
plt.grid()
plt.xlabel("Modulated Signal")
plt.ylabel("Amplitude")
plt.plot(time,Modulated_Signal,color='black')

plt.subplots_adjust(hspace=1)
fig = plt.gcf()
fig.set_size_inches(8,9)
plt.show()
```
